```python
relevant_columns = ('Non-domestic violence related assault',
                    'Robbery without a weapon',
                    'Robbery with a weapon not a firearm',
                    'Robbery with a firearm',
                       )
df_geodata = df_geodata[df_geodata['bcsrcat'].isin(relevant_columns)]
df_geodata = df_geodata.rename(columns={'bcsrgclat':'lat','bcsrgclng':'lon'})

print(df_geodata.head())
df_geodata['time_of_day'] = pd.to_datetime(df_geodata['incsttm'])
print(df_geodata['time_of_day'].head())
```

```python
for i in range(24):
    mask = (df_geodata["time_of_day"] > "2016-07-31 {0}:00:00".format(i)) & (df_geodata["time_of_day"] < "2016-07-31 {0}:59:59".format(i))
    print(i)
    df_geodata_reduced = df_geodata.loc[mask]
    df_geodata_reduced = df_geodata_reduced[['lat','lon']]
    df_geodata_reduced.to_json("{0}_incident.json".format(i),orient="records")
```
